```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
```

```python
df
```

```python
def create_textual_representation(row):
    texttual_representation = f"""Type: {row['type']},
Title: {row['title']},
Director: {row['director']},
Cast: {row['cast']},
Released: {row['release_year']},
Genres: {row['listed_in']},
Description: {row['description']}"""

    return texttual_representation
```

```python
df['textual_representation'] = df.apply(lambda row: create_textual_representation(row), axis=1)
df
```

```python
print(df['textual_representation'].values[0])
```

```python
import faiss
import requests
import numpy as np

dim = 4096

index = faiss.IndexFlatL2(dim)

X = np.zeros((len(df['textual_representation']), dim), dtype=np.float32)
```

```python
for i, representation in enumerate(df['textual_representation']):
    if i % 30 == 0:
        print('Processed', str(i), 'instances')
    
    try:
        res = requests.post('http://localhost:11434/api/embeddings', 
                            json={
                                'model': 'llama2',
                                'prompt': representation
                            })
        res.raise_for_status()
        embedding = res.json().get('embedding', [])
    except requests.RequestException as e:
        print(f"Request failed for instance {i}: {e}")
        embedding = []
    
    if embedding:
        X = np.vstack((X, np.array(embedding)))
    else:
        X = np.vstack((X, np.zeros(dim, dtype=np.float32)))

X = np.array(X)
index.add(X)
```

```python
faiss.write_index(index, 'index')
```
